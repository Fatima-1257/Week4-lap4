```python
total = 0
for i in range(10):
    n = int(input("Enter number: "))
    if n == 0:
        break
    if n < 0:
        continue
    if n % 2 == 0:
        total += n
print(total)
```

```python
x = int(input("Enter a number: "))
for i in range(1, 13):
    res = x * i
    if res > 50:
        break
    if i % 3 == 0:
        continue
    print(x, "*", i, "=", res)
```

```python
import random
secret = random.randint(1, 20)
count = 0
while True:
    g = int(input("Guess: "))
    if g < 1 or g > 20:
        continue
    count += 1
    if g == secret:
        print("Correct")
        print("Attempts:", count)
        break
```

```python
n = int(input("Enter n: "))
c = 0
for i in range(1, n + 1):
    if i > 50:
        break
    if i % 8 == 0:
        continue
    if i % 4 == 0:
        c += 1
print(c)
```

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        if i * j > 6:
            break
        if j == 3:
            continue
        print(j, end=" ")
    print()
```
